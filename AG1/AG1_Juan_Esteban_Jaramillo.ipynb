```
#Quick_sort

#Definimos la lista a ordenar con nuestro algoritmo
A = [9187, 244, 4054, 9222, 8373, 4993, 5265, 5470, 4519, 7182, 2035, 3506, 4337, 7580, 2554, 2824, 8357, 4447, 7379]

#Creamos nuestra función quick_sort
def quick_sort(A):
  if len(A) == 1:
    return A
  if len(A) == 2:
    return [min(A),max(A)]
  
  izq=[]
  Der=[]
  
  #Elegimos nuestro pivote
  pivote = sum(A[0:3])/3 #Para el promedio realicé un sum con recursividad de código
  
  for i in A:
    if i <= pivote:
      izq.append(i)
    else:
      Der.append(i)
  
  return quick_sort(izq) + quick_sort(Der)

#Ejecutamos la función

@calcular_tiempo
def QS(A):
  return quick_sort(A)

print(QS(A))

```

```
#Decorador para calcular_tiempo a la función para la cual queremos calcular el tiempo de ejecución(Isabel Vázquez)

from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f):
   
    def wrapper(*args, **kwargs ):       
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = time() - inicio
        print("Tiempo de ejecución para algoritmo: "+str(tiempo))
        return resultado
    
    return wrapper
from time import time
```

```
SISTEMA = [25, 10, 5, 1]

#Definimos la función de cambio de monedas
@calcular_tiempo
def cambio_monedas(C,SISTEMA):
  SOLUCION = [0 for i in range(len(SISTEMA))]
  
  VALOR_ACUMULADO = 0
  
  for i in range(len(SISTEMA)):
    monedas = int((C - VALOR_ACUMULADO)/SISTEMA[i])
    
    SOLUCION[i] = monedas
    
    VALOR_ACUMULADO += monedas*SISTEMA[i]
    
    if C == VALOR_ACUMULADO:
      return SOLUCION
  


cambio_monedas(1999,SISTEMA)
```

```
N = 4

solucion = [0 for i in range(N)]

etapa = 0

def es_prometedora(solucion, etapa):
  #print(SOLUCION)
  for i in range(etapa+1):
    if solucion.count(solucion[i]) > 1:
      return False
    
    for j in range(i+1, etapa +1):
      
      if abs(i - j) == abs(solucion[i] - solucion[j]):
        return False
    
    return True


def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if solucion[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")
  


def reinas(N,solucion,etapa): #Tamaño del tablero
  
  for i in range(1,N+1):
    solucion[etapa] = i
    
    #print(solucion)
    if es_prometedora(solucion,etapa):
      if etapa == N-1 :
        print("\n\nLa solución es:")
        print(solucion)
        escribe(solucion)
      else:
        #print("Es prometedora\n###################")
        reinas(N,solucion,etapa+1)
    else:
      #print("NO PROMETEDORA\n###################")
      None
      
    solucion[etapa] = 0
    

reinas(N,solucion,etapa)
```
